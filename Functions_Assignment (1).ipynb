1. What is the difference between a function and a method in Python?

In Python, functions and methods are both blocks of reusable code, but they differ in how they are called and their association with objects.

Function:
1. A function is a standalone block of code that performs a specific task.
2. It is defined using the def keyword followed by a function name, parameters (optional), and a colon.
3. It is called directly by its name, passing any required arguments.
4. It does not operate on a specific object or class instance.

For Example:

  def add(x, y):

      return x + y

  result = add(5, 3)

  print(result)
 # Output: 8

Method:
1. A method is a function that is associated with an object of a class.
2. It is defined within a class and is called using the object name followed by a dot and the method name.
3. It implicitly receives the object instance as its first argument (conventionally named self).
4. It can access and modify the object's data (attributes).

For Example:

class Dog:

    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        print("Woof!")

    def describe(self):
        print(f"This is a {self.breed} named {self.name}")

my_dog = Dog("Buddy", "Golden Retriever")

my_dog.bark()
     
my_dog.describe()

# Output: Woof!
# Output: This is a Golden Retriever named Buddy

Parameters:

The different ways to define and call a function in Python are:

(A). Defining a Function

1. Use the def keyword followed by the function name and parentheses ().
2. Parameters can be placed inside the parentheses, separated by commas.
3. A colon : ends the function definition line.

1. Assign default values to parameters in the function definition.
2. If an argument isn't provided during the function call, the default value is used.

For Example:

def power(base, exponent=2):

   print(power(3))

4. What is the purpose of the `return` statement in a Python function?

The return statement in a Python function serves two primary purposes:

1. Exiting the function:

When a return statement is encountered, the function's execution immediately terminates, and control is transferred back to the caller. Any code after the return statement within the function will not be executed.

2. Returning a value:

The return statement can optionally specify a value or an object that the function will send back to the caller. This returned value can be used for further processing or assigned to a variable. If no value is explicitly specified, the function implicitly returns None.

# An example demonstrating the use of the return statement:

def add_numbers(x, y):

  """This function adds two numbers and returns the result."""

  result = x + y

  return result

sum_result = add_numbers(5, 3)

print(sum_result)  

In this example:

1. The add_numbers function takes two arguments, x and y.
2. It calculates their sum and stores it in the result variable.
3. The return result statement sends the calculated sum back to the caller.
4. The caller receives the returned value and assigns it to the sum_result variable.
5. The print statement displays the returned value.

6. Explain the concept of generators in Python and how they are defined.

A generator function is defined using the def keyword, just like a regular function. However, instead of using return to send back a value, it uses yield to produce a value. When yield is encountered, the function pauses, and the yielded value is returned. When the next value is requested (e.g., by calling next() on the generator object or iterating over it in a for loop), the function resumes from where it left off, continuing until the next yield or the end of the function is reached.

For Example:

def my_generator(n):

    """Generates numbers from 0 up to n-1."""
    i = 0
    while i < n:
        yield i
        i += 1

# Creating a generator object
generator = my_generator(5)

# Iterating over the generator
for number in generator:
    print(number)

Explanation:

1. def my_generator(n): defines a generator function.
2. i = 0 initializes a counter.
3. while i < n: loop continues until the counter reaches n.
4. yield i produces the current value of i.
5. i += 1 increments the counter.
6. generator = my_generator(5) creates a generator object by calling the function.
7. for number in generator: iterates over the generator.
8. print(number) prints the yielded value.

7. What are the advantages of using generators over regular functions?

Generators offer several advantages over regular functions, especially when dealing with large datasets or infinite sequences. Here are some key benefits:

(A). Memory Efficiency:

1. Generators produce values on demand using the yield keyword, rather than storing all values in memory like a regular function. This makes them more memory-efficient, especially with large datasets.
2. Regular functions compute all values upfront and store them in memory, which can be inefficient for large datasets.

(B). Lazy Evaluation:

1. Generators compute values only when needed, which can improve performance and responsiveness, especially when the entire sequence of values is not required at once.
2. Regular functions compute all values at once, whether or not they are all needed.

(C). Simplified Iterative Logic:

1. Generators make it easier to write and maintain code for producing sequences, reducing the need for complex state management.
2. Regular functions often require more complex code to manage the state of iteration.

(D). Flexibility:

1. Generators can be used to create infinite sequences, which is ideal for real-time data processing.
2. Regular functions cannot easily represent infinite sequences.

Example:

# Regular function

def regular_function(n):

    result = []
    for i in range(n):
        result.append(i * 2)
    return result

print("Regular function",regular_function(5))

# Generator function
def generator_function(n):

    for i in range(n):
        yield i * 2

gen = generator_function(5)

print("Generator function",list(gen))

8. What is a lambda function in Python and when is it typically used?

A lambda function in Python is a small, anonymous function defined using the lambda keyword. It's designed for short, simple operations and is often used as an argument to other functions like map(), filter(), and sort(). Lambda functions can take any number of arguments but can only have one expression.

Syntax:

lambda arguments: expression

Example:

# A lambda function that squares a number
square = lambda x: x**2

# Use the lambda function
result = square(5)

print(result)  # Output: 25

# Use lambda with map() to square each number in a list
numbers = [1, 2, 3, 4, 5]

squared_numbers = list(map(lambda x: x**2, numbers))

print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

# When to use lambda functions:

1. Simple operations:
When you need a function for a short-term, one-time use, especially for simple tasks.
2. With higher-order functions:
When passing a function as an argument to another function like map(), filter(), or sorted().
3. Reducing code clutter:
When you need a small function inline within a larger piece of code, without needing to name it separately.

9. Explain the purpose and usage of the `map()` function in Python.

The map() function in Python is a built-in function that applies a given function to each item of an iterable (like a list, tuple, or string) and returns an iterator of the results. This allows for efficient and concise code when you need to perform the same operation on multiple elements.

Purpose:

1. To apply a specific function to every element of an iterable.
2. To transform the elements of an iterable into a new iterable.
3. To avoid explicit loops when processing iterables.

Usage:

The basic syntax is:

map(function, iterable, ...)
1. function: The function to apply to each element.
2. iterable: The iterable to process.
3. ... : Additional iterables if the function takes more than one input.

Example:

def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]

squared_numbers = map(square, numbers)
# returns a map object, which is an iterator

squared_numbers_list = list(squared_numbers)
# converts map object to a list

print(squared_numbers_list)  

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

map(), filter(), and reduce() are higher-order functions in Python that operate on iterables (like lists, tuples, etc.) and return a new iterable or a single value.

 A breakdown of their differences:

1. map()

(1). Purpose: Applies a given function to each item in an iterable and returns a new iterator containing the results.

(2). Output: Returns a map object (an iterator), which can be converted to a list, tuple, etc.

(3). Use case: Transforms each element of a collection.

(4). Example:

numbers = [1, 2, 3, 4]

     squared_numbers = map(lambda x: x**2, numbers)
     print(list(squared_numbers))  # Output: [1, 4, 9, 16]

2. reduce()  

(A). Purpose: Applies a function cumulatively to the items of an iterable, reducing it to a single value.

(B). Output: Returns a single value, not an iterable.

(C). Use case: Aggregates elements of a collection into a single result.

Note: reduce() is part of the functools module in Python

(D). Example:

 from functools import reduce

     numbers = [1, 2, 3, 4]
     sum_of_numbers = reduce(lambda x, y: x + y, numbers)
     print(sum_of_numbers)  

3. filter()

(A). Purpose: Filters elements from an iterable based on a condition specified by a function.

(B). Output: Returns a filter object (an iterator) containing only the elements for which the function returns True.

(C). Use case: Selects elements from a collection that meet a specific condition.

(D). Example:

 numbers = [1, 2, 3, 4, 5, 6]

     even_numbers = filter(lambda x: x % 2 == 0, numbers)
     print(list(even_numbers))

```python
def sum_even_numbers(numbers):
  return sum(num for num in numbers if num % 2 == 0)
my_list = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(my_list)
print(result)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
  if n < 2 or n > 200:
    return False
  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      return False
  return True

  print(is_prime(7))      #True
  print(is_prime(100))    #False
  print(is_prime(199))    #True
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
  def __init__(self, max_terms):
    self.max_terms = max_terms
    self.count = 0
    self.a, self.b = 0, 1

  def __iter__(self):
    return self

  def __next__(self):
    if self.count >= self.max_terms:
      raise StopIteration
    if self.count == 0:
       self.count += 1
       return 0
    elif self.count == 1:
         self.count += 1
         return 1
    else:
      self.a, self.b = self.b, self.a + self.b
      self.count += 1
      return self.b

fib = FibonacciIterator(10)
for num in fib:
  print(num, end='')
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
  for exponent in range(max_exponent + 1):
    yield 2 ** exponent
for value in powers_of_two(5):
  print(value)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def file_line_generator(file_path):
  try:
    with open(file_path, 'r') as file:
      for line in file:
        yield line.strip()
  except FileNotFoundError:
        print("File not Found.")

for line in file_line_generator('example.txt'):
    print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))
print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
input_str = "Hello, how are you?"
vowels = "aeiouAEIOU"
no_vowels = ''.join(filter(lambda ch: ch not in vowels, input_str))
print("String without vowels:", no_vowels)
```
